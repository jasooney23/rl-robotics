```python
import stable_baselines3 as baselines, gymnasium as gym
```

```python
env = gym.make("Reacher-v5")
model = baselines.SAC("MlpPolicy", env, verbose=1)
model.learn(total_timesteps=1e6)
```

```python
model.save("sac_Reacher_sb3")
```

```python
env = gym.make("Reacher-v5", render_mode="human")
while True:
    obs, info = env.reset()
    for _ in range(1000):
        action, _states = model.predict(obs, deterministic=True)
        obs, rewards, dones, trunc, info = env.step(action)
        env.render()
```
